```python
# --- Import Required Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
from sklearn.metrics import mean_squared_log_error, median_absolute_error, max_error, mean_poisson_deviance, mean_gamma_deviance
import joblib

# --- Load Dataset ---
data = pd.read_csv('p2_sorted_by_load.csv')
data.dropna(inplace=True)

# --- Encode Load Condition ---
label_encoder = LabelEncoder()
data['Load Condition'] = label_encoder.fit_transform(data['Load Condition'])

# --- Define Constants ---
FARADAY_CONSTANT = 96485  # C/mol
STACK_RATED_POWER = 100  # W
NUMBER_OF_CELLS = 20
STACK_VOLTAGE = 12  # V
EFFECTIVE_AREA_PEM_CELL = 26  # cmÂ²
TIME_CONVERSION = 60  # sec/min
MOLE_TO_VOLUME_CONVERSION = 23.65  # L/mol
REFERENCE_VOLTAGE_HEATING = 1.25  # V
MAXIMUM_FUEL_CELL_VOLTAGE = 1.48  # V
AVERAGE_CELL_VOLTAGE = STACK_VOLTAGE / NUMBER_OF_CELLS

# --- Calculate Target Columns ---
data['Power Output'] = data['Voltage (V)'] * data['Current (A)']
data['Hydrogen Consumption Rate'] = (data['Power Output'] / (2 * AVERAGE_CELL_VOLTAGE * FARADAY_CONSTANT)) * TIME_CONVERSION * MOLE_TO_VOLUME_CONVERSION
data['Oxygen Consumption Rate'] = (data['Power Output'] / (4 * AVERAGE_CELL_VOLTAGE * FARADAY_CONSTANT)) * TIME_CONVERSION * MOLE_TO_VOLUME_CONVERSION
data['Water Production'] = (data['Power Output'] / (2 * AVERAGE_CELL_VOLTAGE * FARADAY_CONSTANT)) * TIME_CONVERSION * MOLE_TO_VOLUME_CONVERSION
data['Heat Generation Rate'] = data['Power Output'] * ((REFERENCE_VOLTAGE_HEATING / AVERAGE_CELL_VOLTAGE) - 1)
data['Power Density'] = data['Power Output'] / EFFECTIVE_AREA_PEM_CELL
data['Current Density'] = data['Current (A)'] / EFFECTIVE_AREA_PEM_CELL
data['Efficiency'] = (AVERAGE_CELL_VOLTAGE / MAXIMUM_FUEL_CELL_VOLTAGE) * 100

# --- Define Target Columns and Features ---
target_columns = ['Power Output', 'Efficiency', 'Hydrogen Consumption Rate', 'Oxygen Consumption Rate',
                  'Water Production', 'Heat Generation Rate', 'Power Density', 'Current Density']
X = data[['Voltage (V)', 'Current (A)']]  # Features for prediction
y_load = data['Load Condition']  # Load condition as target variable
y_targets = data[target_columns]  # Target columns for prediction

# --- Split Data ---
X_train, X_test, y_train_load, y_test_load = train_test_split(X, y_load, test_size=0.2, random_state=42)
X_train_tgt, X_test_tgt, y_train_targets, y_test_targets = train_test_split(X, y_targets, test_size=0.2, random_state=42)

# --- Scale Features ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --- Train Random Forest for Load Condition ---
rf_load = RandomForestRegressor(n_estimators=100, random_state=42)
rf_load.fit(X_train_scaled, y_train_load)

# --- Generate Synthetic Data for Higher Loads ---
additional_loads = [5, 6, 7, 8, 9, 10]  # New load conditions
synthetic_data = []

# Generate realistic voltage and current combinations for higher loads
voltage_range = np.linspace(X['Voltage (V)'].min(), X['Voltage (V)'].max(), 50)
current_range = np.linspace(X['Current (A)'].min(), X['Current (A)'].max(), 50)

for load in additional_loads:
    for v in voltage_range:
        for c in current_range:
            synthetic_data.append({'Voltage (V)': v, 'Current (A)': c, 'Load Condition': load})

# --- Create Synthetic DataFrame ---
synthetic_df = pd.DataFrame(synthetic_data)

# --- Calculate Target Values for Synthetic Data ---
synthetic_df['Power Output'] = synthetic_df['Voltage (V)'] * synthetic_df['Current (A)']
synthetic_df['Hydrogen Consumption Rate'] = (synthetic_df['Power Output'] / (2 * AVERAGE_CELL_VOLTAGE * FARADAY_CONSTANT)) * TIME_CONVERSION * MOLE_TO_VOLUME_CONVERSION
synthetic_df['Oxygen Consumption Rate'] = (synthetic_df['Power Output'] / (4 * AVERAGE_CELL_VOLTAGE * FARADAY_CONSTANT)) * TIME_CONVERSION * MOLE_TO_VOLUME_CONVERSION
synthetic_df['Water Production'] = (synthetic_df['Power Output'] / (2 * AVERAGE_CELL_VOLTAGE * FARADAY_CONSTANT)) * TIME_CONVERSION * MOLE_TO_VOLUME_CONVERSION
synthetic_df['Heat Generation Rate'] = synthetic_df['Power Output'] * ((REFERENCE_VOLTAGE_HEATING / AVERAGE_CELL_VOLTAGE) - 1)
synthetic_df['Power Density'] = synthetic_df['Power Output'] / EFFECTIVE_AREA_PEM_CELL
synthetic_df['Current Density'] = synthetic_df['Current (A)'] / EFFECTIVE_AREA_PEM_CELL
synthetic_df['Efficiency'] = (AVERAGE_CELL_VOLTAGE / MAXIMUM_FUEL_CELL_VOLTAGE) * 100

# --- Merge Synthetic and Original Data ---
final_data = pd.concat([data, synthetic_df], ignore_index=True)

# --- Redefine Features and Targets After Merging ---
X_final = final_data[['Voltage (V)', 'Current (A)']]
y_final_load = final_data['Load Condition']
y_final_targets = final_data[target_columns]

# --- Split Merged Data for Retraining ---
X_train_final, X_test_final, y_train_load_final, y_test_load_final = train_test_split(X_final, y_final_load, test_size=0.2, random_state=42)
X_train_tgt_final, X_test_tgt_final, y_train_targets_final, y_test_targets_final = train_test_split(X_final, y_final_targets, test_size=0.2, random_state=42)

# --- Scale Merged Features ---
X_train_final_scaled = scaler.fit_transform(X_train_final)
X_test_final_scaled = scaler.transform(X_test_final)

# --- Retrain Random Forest for Load Condition on Merged Data ---
rf_load_final = RandomForestRegressor(n_estimators=100, random_state=42)
rf_load_final.fit(X_train_final_scaled, y_train_load_final)

# --- Retrain Random Forest for Target Variables on Merged Data ---
rf_target_final = RandomForestRegressor(n_estimators=100, random_state=42)
rf_target_final.fit(X_train_final_scaled, y_train_targets_final)

# --- Save Retrained Models ---
joblib.dump(rf_load_final, 'rf_load_model.pkl')
joblib.dump(rf_target_final, 'rf_target_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
```

```python
# --- Define Overall Evaluation Metrics ---
def evaluate_overall_metrics(y_true, y_pred):
    y_true_clipped = np.clip(y_true, a_min=1e-10, a_max=None)
    y_pred_clipped = np.clip(y_pred, a_min=1e-10, a_max=None)

    print(f"--- Overall Evaluation Metrics ---")
    print(f"Mean Absolute Error: {mean_absolute_error(y_true, y_pred):.4f}")
    print(f"Mean Squared Error: {mean_squared_error(y_true, y_pred):.4f}")
    print(f"Root Mean Squared Error: {np.sqrt(mean_squared_error(y_true, y_pred)):.4f}")
    print(f"R2 Score: {r2_score(y_true, y_pred):.4f}")
    print(f"Explained Variance Score: {explained_variance_score(y_true, y_pred):.4f}")
    print(f"Max Error: {max_error(y_true, y_pred):.4f}")
    print(f"Mean Gamma Deviance: {mean_gamma_deviance(y_true_clipped, y_pred_clipped):.4f}")
    print("-" * 40)

# --- Concatenate All True and Predicted Values for Overall Metrics ---
y_true_combined = np.concatenate([y_test_targets_final[col].values for col in target_columns])
y_pred_combined = np.concatenate([y_pred_targets[:, i] for i in range(len(target_columns))])

# --- Evaluate Overall Metrics ---
evaluate_overall_metrics(y_true_combined, y_pred_combined)
```
